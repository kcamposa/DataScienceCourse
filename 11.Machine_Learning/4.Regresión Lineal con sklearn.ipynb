# **Regresion Lineal con Scikit-Learn**

```python
from sklearn import linear_model
```

**1.Definir las variables dependientes e independientes**

```python
y = df_boston['Value']
X = df_boston[['Rooms', 'Distance']]
```

**2.Ajustar modelo**

```python
lm = linear_model.LinearRegression()
lm.fit(X, y) # .fit(indep, dep)
```

**3.Predecir los valores**

```python
lm.predict(X)[:5]
```

**4.Tabla de Regresion**

```python
# r2 score : R-squared
lm.score(X, y)
```

```python
# coeficiente de Rooms y Distance
lm.coef_
```

```python
# intercepto : constante
lm.intercept_
```
