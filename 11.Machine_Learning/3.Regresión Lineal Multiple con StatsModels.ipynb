# **Regresion Lineal Multiple**

```python
import statsmodels.api as sm
```

**1.Definir las variables dependientes e independientes**

```python
y = df_boston['Value'] # variable dependiente

X = df_boston[['Rooms', 'Distance']] # variable independiente
```

**2.Crear la regresion: Agregar la constante y ajustar el modelo**

```python
X = sm.add_constant(X) # agregamos la constante
lm = sm.OLS(y, X).fit()
```

```python
lm.summary()
```

```python
# Coef Rooms : 8.8014 : b1
# Coef const : -34.6361 : b0
# Coef Distance : 0.4888 : b2

# mult lineal ecu : y = b1x1 + b0 + b2x2 + ...
y_pred = 8.8014*X['Rooms'] + 0.4888*X['Distance'] - 34.6361 # 
```
