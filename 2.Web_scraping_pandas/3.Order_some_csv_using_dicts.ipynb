```python
import pandas as pd
```

```python
dict_countries = {
    'Spanish La Liga':'SP1', 'Spanish Segunda Division':'SP2',
    'German Bundesliga':'D1',
    'English Premier League':'E0', 'English League 1':'E2', 'English League 2':'E3',
}
```

```python
dict_countries['Spanish La Liga']
```

```python
root = 'https://www.football-data.co.uk/mmz4281/'
```

```python
dict_historical_data = {}

for league in dict_countries:
  frames = []
  for season in range(15, 21):
    df = pd.read_csv(root + str(season) + str(season+1) + '/' + dict_countries[league] + '.csv') # after '.csv' if will show you an error, write:     , encoding='unicode_scape'
    df.insert(1, 'season', season)
    frames.append(df)
  
  df_concat = pd.concat(frames)
  dict_historical_data[league] = df_concat
```
