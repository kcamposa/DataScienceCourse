**Read some CSV files at the same time from a page**

```python
import pandas as pd
```

```python
# https://www.football-data.co.uk/mmz4281/2223/E0.csv

# https://www.football-data.co.uk/mmz4281/2223/E1.csv

# https://www.football-data.co.uk/mmz4281/2223/E2.csv

# https://www.football-data.co.uk/mmz4281/2223/E3.csv

# https://www.football-data.co.uk/mmz4281/2223/EC.csv
```

```python
root = 'https://www.football-data.co.uk/mmz4281/'
```

```python
leagues = ['E0', 'E1', 'E2', 'E3', 'EC']
frames = []

for league in leagues:
  df = pd.read_csv(root + '2223' + '/' + league + '.csv') # after '.csv' if will show you an error, write:     , encoding='unicode_scape'
  frames.append(df)
```

```python
# size of frames
len(frames)
```

```python
# show specific frame
frames[4].head()
```

**Multiple Seasons**

```python
leagues = ['E0', 'E2', 'E3']
frames = []

for league in leagues:
  for season in range(15, 21):
    df = pd.read_csv(root + str(season) + str(season+1) + '/' + league + '.csv') # after '.csv' if will show you an error, write:     , encoding='unicode_scape'
    df.insert(1, 'season', season)
    frames.append(df)
```

```python
len(frames)
```
