**Split-Apply-Combine STRATEGY**

```python
df_cars['Vehicle_type'].value_counts()
```

```python
# split
car_filter = df_cars['Vehicle_type']=='Car'
passeger_filter = df_cars['Vehicle_type']=='Passenger'
```

```python
# apply
car_average = round(df_cars[car_filter]['Sales_in_thousands'].mean(), 2)
passeger_average = round(df_cars[passeger_filter]['Sales_in_thousands'].mean(), 2)
```

```python
# combine
df_combined = pd.DataFrame({'Vehicle_type': ['Car', 'Passenger'],
                            'Sales_in_thousands': [car_average, passeger_average]})
df_combined
```
