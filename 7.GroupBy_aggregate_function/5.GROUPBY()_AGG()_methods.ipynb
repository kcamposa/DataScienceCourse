```python
import pandas as pd
import numpy as np
```

```python
dir_datasets = '/content/drive/MyDrive/DataScience/Data Science Course/0.Datasets/{}'
df_cars = pd.read_csv(dir_datasets.format('Car_sales.csv'), encoding='latin-1')
```

```python
df_cars = df_cars[['Manufacturer', 'Sales_in_thousands', 'Vehicle_type', 'Price_in_thousands', 'Engine_size', 'Horsepower', 'Fuel_capacity']]
df_cars
```

# **GROUPBY() AND AGG()**

```python
df_cars.columns
```

```python
df_cars.groupby('Vehicle_type').agg(['min', 'max'])
```

```python
df_cars.groupby('Vehicle_type', as_index=False).agg(min_engine_size=('Engine_size', 'min'),
                                                    max_engine_size=('Engine_size', 'max'),
                                                    min_horsepower=('Horsepower', 'min'),
                                                    max_horsepower=('Horsepower','max'))
```

```python
df_cars.groupby('Manufacturer').agg(sum_sales =('Sales_in_thousands', 'sum'),
                                    average_price=('Price_in_thousands','mean')).head()
```

```python
df_cars.groupby('Manufacturer').sum()['Sales_in_thousands'].head()
```

```python
df_cars.groupby('Manufacturer').mean()['Price_in_thousands'].head()
```
