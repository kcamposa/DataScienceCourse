# **.AGG() METHOD**

The agg() method allows you to apply a function or a list of function names to be executed along one of the axis of the DataFrame, default 0, which is the index (row) axis.

**sum, min, max, mea ...**

**Note:** the agg() method is an alias of the aggregate() method.

```python
df_cars.agg('sum')
```

```python
df_cars.agg('mean')
```

```python
df_cars.agg(['sum', 'mean'])
```

```python
df_cars.agg({'Sales_in_thousands':['sum', 'mean'],
             'Price_in_thousands':['sum','max']})
```

```python
# we are doing a sum between Sales_in_thousands and Price_in_thousands 
df_cars[['Sales_in_thousands', 'Price_in_thousands']].agg('sum', axis=1) 

# result:
```

```python
# another example is
df_cars['Sales_in_thousands'] + df_cars['Price_in_thousands']
```

```python
df_cars.agg(x=('Sales_in_thousands', 'sum'), 
            y=('Price_in_thousands', 'sum'))
```
