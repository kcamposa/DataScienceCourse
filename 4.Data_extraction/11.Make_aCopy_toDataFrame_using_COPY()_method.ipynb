```python
import pandas as pd
import numpy as np
```

```python
dir_datasets = '/content/drive/MyDrive/DataScience/Data Science Course/0.Datasets/{}'
df_players = pd.read_csv(dir_datasets.format('players_20.csv'), sep=',')
```

```python
df_players = df_players[['short_name', 'long_name',	'age',	'dob',	'height_cm',	'weight_kg',	'nationality',	'club']]
```

```python
df_players.set_index('short_name', inplace=True)
```

```python
df_players.head()
```

# **COPY()**

### **deep = True**

```python
df_players_copy = df_players.copy()
```

```python
df_players.loc['L. Messi', 'height_cm'] = 180
```

```python
df_players.head(1)
```

```python
df_players_copy.head(1)
```

### **deep = False**

```python
df_players_shallow_copy = df_players.copy(deep=False)
```

```python
df_players.loc['Cristiano Ronaldo', 'height_cm'] = 190
```

```python
df_players.head(2)
```

```python
df_players_shallow_copy.head(2)
```
