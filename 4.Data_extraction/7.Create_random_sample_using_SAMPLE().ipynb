# **SAMPLE()**
With sample you can get a sample of dataframe 

```python
# get 10 randoms elements of nationality column
df_players['nationality'].sample(10, random_state=50) # random_state = keep the same random result
```

```python
# get 20% of random samples of dataframe
df_players.sample(frac=0.2, random_state=999)
```

```python
# increment the sample
df_players.sample(frac=2, replace=True)
```
