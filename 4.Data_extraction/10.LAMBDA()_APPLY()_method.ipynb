# **LAMBDA FUNCTION**

```python
# basic function
def sum_values(a, b):
  return  a + b
  
sum_values(2, 3)
```

```python
# lambda function
sum_values_lambda = lambda a,b:a+b

sum_values_lambda(2,3)
```

# **APPLY + LAMBDA FUNCTION**

```python
# use lambda for convert height_cm to meters
df_players['height_cm'].apply(lambda x: x/100)

# alternative
df_players['height_cm']/100
```

```python
# use the lambda function for convert long_name to uppercase
df_players['long_name'].apply(lambda x: x.upper())

# alternative
df_players['long_name'].str.upper()
```

```python
# use the lamda function to get the year of "dob"
df_players['dob'] = df_players['dob'].astype('datetime64')

df_players['dob'].apply(lambda x: x.year)

# alternative
df_players['dob'].dt.year
```

```python
# use the lambda function for the dataframe to get IMC
df_players.apply(lambda x: round(x['weight_kg'] / ((x['height_cm']/100)**2),2), axis=1)
```
