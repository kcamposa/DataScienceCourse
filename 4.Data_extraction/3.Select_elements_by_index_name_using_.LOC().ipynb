```python
df_data_new_players = df_players.loc['L. Messi':'M. Salah', columns]
```

```python
# for a new dataframe
ages = []
for age in df_data_new_players['dob']:
  #print(int(age[:4]))
  ages.append(2023-int(age[:4]))
```

```python
df_data_new_players['current_age'] = ages
```

```python
df_data_new_players
```

```python
# for a current dataframe
ages = []
for age in df_players['dob']:
  #print(int(age[:4]))
  ages.append(2023-int(age[:4]))
```

```python
df_players['age_in_2023'] = ages
```

**SELECT ELEMENTS USING CONDITIONS**

```python
# one condition : select size players taller than 180 cm
columns = ['age',	'dob',	'height_cm',	'weight_kg'	]

df_players.loc[df_players['height_cm']>180, columns]
```

```python
# multiple conditions : select size players taller than 180 cm and they have to be from Argentina

df_players.loc[(df_players['height_cm']>180) & (df_players['nationality']=='Argentina'), :]
```
