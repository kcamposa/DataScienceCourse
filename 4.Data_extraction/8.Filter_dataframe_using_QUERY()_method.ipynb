```python
import pandas as pd
import numpy as np
```

```python
dir_datasets = '/content/drive/MyDrive/DataScience/Data Science Course/0.Datasets/{}'
df_players = pd.read_csv(dir_datasets.format('players_20.csv'), sep=',')
```

```python
df_players = df_players[['short_name', 'long_name',	'age',	'dob',	'height_cm',	'weight_kg',	'nationality',	'club']]
```

```python
df_players.set_index('short_name', inplace=True)
```

```python
df_players.head()
```

# **QUERY()**

```python
df_players.query('age>34')

df_players[df_players['age']>34]
```

```python
# select players from Italy and older than 34
df_players.query('age>34 and nationality=="Italy"')

# boolean slicing example
df_players[(df_players['age']>34) & (df_players['nationality']=='Italy')].head()
```

```python
# negative operator for the first example
df_players.query('not(age>34)')

# boolean slicing example
df_players[~(df_players['age']>34)]
```

```python
# convert the height from cm to meters and select players taller than 1.80
df_players.query("height_cm/100 > 1.8")

# boolean slicing example
df_players[((df_players['height_cm']/100)>1.8)]
```

```python
# select players born before 1990
df_players.dtypes

df_players['dob'] = df_players['dob'].astype('datetime64')

df_players.query("dob>1990")

df_players.query("dob.dt.year > 1990")
```

```python
# boolean slicing example
df_players[df_players['dob'].dt.year>1990]
```
