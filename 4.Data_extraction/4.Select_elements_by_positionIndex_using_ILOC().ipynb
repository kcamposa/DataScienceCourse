```python
import pandas as pd
import numpy as np
```

```python
dir_datasets = '/content/drive/MyDrive/DataScience/Data Science Course/0.Datasets/{}'
df_players = pd.read_csv(dir_datasets.format('players_20.csv'), sep=',')
```

```python
df_players.set_index('short_name')
```

```python
df_players = df_players[['long_name',	'age',	'dob', 'height_cm',	'weight_kg',	'nationality',	'club']]
df_players.head()
```

# **SELECT BY VALUES**

```python
# get the Messi height_cm
df_players.iloc[0, 3]
```

```python
# get the Cristiano Ronaldo height_cm
df_players.iloc[1, 3]
```

```python
# get all rows of height_cm
df_players.iloc[:, 3]
```

# **SELECT BY LIST VALUES**

```python
# get data about L. Messi and Cristiano Ronaldo
df_players.iloc[0:2, :]
```

```python
# get the L. Messi's and Cristiano Ronaldo's height_cm
df_players.iloc[0:2, 3]
```

```python
# get the L. Messi's height_cm and weight_kg\
df_players.iloc[0, 3:5]
```

```python
# get the height_cm and weight_kg of L. Messi and Cristiano Ronaldo
df_players.iloc[0:2, 3:5]
```

# **SELECT A RANGE OF DATA WITH A SLICE**

```python
columns = [1, 3, 5]

df_players.iloc[0, columns]
```

# **SELECT WITH CONDITIONS**

```python
columns = [0, 2, 5]

df_players.iloc[list(df_players['height_cm']>180), columns]
```

```python
df_players.iloc[list((df_players['height_cm']>180) & (df_players['nationality']=='Argentina')), columns]
```
