**ORDER A DATAFRAME**

```python
# order a column
df_exams.sort_values('math score')
```

```python
# order column desc
df_exams.sort_values('math score', ascending=False)
```

```python
# order multiple column desc
df_exams.sort_values(['math score', 'reading score'], ascending=False)
```

```python
# order multiple column desc and udpate dataframe
df_exams.sort_values(['math score', 'reading score'], ascending=False,
                     inplace=True) # inplace update the original dataframe

df_exams
```

```python
# desc order using a key function
df_exams.sort_values('race/ethnicity', ascending=True, 
                     key=lambda col:col.str.lower())
```
