```python
import pandas as pd
import cufflinks as cf
from IPython.display import display,HTML

cf.set_config_file(sharing='public', theme='white', offline=True) # cufflinks

```

```python
cf.getThemes()
```

```python
dir_datasets = '/content/drive/MyDrive/DataScience/Data Science Course/0.Datasets/{}'
df_population_raw = pd.read_csv(dir_datasets.format('population_total.csv'), encoding='latin-1')
```

```python
df_population_raw.head()
```

```python
df_population_raw = df_population_raw.dropna() # remove nulls of dataframe
```

```python
df_population_raw['year'] = df_population_raw['year'].astype('int64')
```

```python
df_population_raw = df_population_raw.pivot(index='year',
                        columns='country',
                        values='population')

df_population_raw = df_population_raw[['United States', 'India', 'China', 'Indonesia', 'Brazil']]

df_population_raw.head()
```

# **INTERACTIVE LINEPLOT**

```python
df_population_raw.iplot(kind='line',
                        xTitle='YEAR',
                        yTitle='POPULATION',
                        title='Year vs Population')
```

# **BARPLOT**

```python
df_population_raw_2020 = df_population_raw[df_population_raw.index.isin([2020])]
df_population_raw_2020 = df_population_raw_2020.T

df_population_raw_2020.iplot(kind='bar',
                        xTitle='COUNTRIES',
                        yTitle='POPULATION',
                        title='Year vs Population')
```

# **MULTIPLE BARPLOT**

```python
df_population_raw_sample = df_population_raw[df_population_raw.index.isin([1980, 1990, 2000, 2010, 2020])]

df_population_raw_sample.iplot(kind='bar',
                        xTitle='COUNTRIES',
                        yTitle='POPULATION',
                        title='Year vs Population')
```

# **BOXPLOT**

```python
df_population_raw['United States'].iplot(kind='box')
```

# **HISTOGRAM**

```python
df_population_raw['United States'].iplot(kind='hist')
```

```python
df_population_raw[['United States', 'Indonesia']].iplot(kind='hist')
```

# **PIECHART**

```python
df_population_raw_2020 = df_population_raw_2020.reset_index()
```

```python
df_population_raw_2020 = df_population_raw_2020.rename(columns={2020:'2020'})
```

```python
df_population_raw_2020.iplot(kind='pie', labels='country', values='2020')
```

# **SCATTER PLOT**

```python
df_population_raw.iplot(kind='scatter',
                        mode='markers',
                        xTitle='YEAR',
                        yTitle='POPULATION',
                        title='Year vs Population')
```
