```python
import pandas as pd
import numpy as np
```

```python
dir_datasets = '/content/drive/MyDrive/DataScience/Data Science Course/0.Datasets/{}'
df_netflix = pd.read_csv(dir_datasets.format('netflix_titles.csv'))
```

```python
df_netflix.head()
```

```python
# le pasamos un cero en texto porque luego necesitamos pasar ese cero a numero, 
# si le pasamos directamente un cero en numero lo calificara como NaN
df_netflix['duration'].fillna('0', inplace=True)
```

```python
# Extraer datos datos de la columna "duration" usando split() y extract()
df_movies = df_netflix[df_netflix['type'] == 'Movie']

df_movies = df_movies.assign(minutes=df_movies['duration'].str.split(expand=True)[0].astype(int))
#df_movies['minutes'] = df_movies['duration'].str.split(expand=True)[0].astype(int) # expand = crea una columna a cada separacion que hizo
```

# **Reemplazar cadenas de texto con replace() o sub()**

```python
import re

# eliminar signos de puntuacion con el metodo str.replace() y regex, el str solo es para cadenas de texto
df_movies['title'].str.replace('[^\w\s]', '', regex=True)

# eliminar puntuacion con el metodo replace() y regex(), aca el replace tambien puede modificar data numerica
df_movies['title'].replace('[^\w\s]', '', regex=True)

# diferencia entre str.replace() y replace()
df_movies['title'].replace(1, 2, regex=False)
df_movies['title'].str.replace(1, 2, regex=False) # aca nos daria error

# eliminar puntuacion con la funcion re.sub() con el metodo apply
df_movies['title'].apply(lambda x:re.sub('[^\w\s]', '', x))
```
