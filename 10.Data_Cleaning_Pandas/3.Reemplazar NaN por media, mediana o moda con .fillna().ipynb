```python
import pandas as pd
import numpy as np
```

```python
dir_datasets = '/content/drive/MyDrive/DataScience/Data Science Course/0.Datasets/{}'
df_netflix = pd.read_csv(dir_datasets.format('netflix_titles.csv'))
```

```python
df_netflix.head()
```

```python
for column in df_netflix.columns:
  percentage = df_netflix[column].isnull().mean()
  print(column + ' : ' + str(round(percentage*100, 2)))
```

```python
df_netflix[df_netflix['rating'].isnull()]
```

```python
# media o mediana cuando son datos numericos
# moda cuando es data categorica
moda = ''.join(df_netflix['rating'].mode()) # .mean()
df_netflix['rating'].fillna(moda, inplace=True)
```

```python
# reemplazar los valores nulos por numero arbitrario, con 0
df_netflix['duration'].fillna(0, inplace=True)
```

```python
# reemplazar numero arbitrario con el numero siguiente o anterior de la fila
df_netflix.fillna(method='bfill') # bfill o ffill : backward fill o forward fill
```
