```python
import pandas as pd
import numpy as np
```

```python
dir_datasets = '/content/drive/MyDrive/DataScience/Data Science Course/0.Datasets/{}'
df_laptops = pd.read_csv(dir_datasets.format('laptop_price.csv'), encoding='latin-1')
```

```python
df_laptops.head()
```

```python
conditions = [
    df_laptops['Price_euros'] > 3000,
    (df_laptops['Price_euros'] > 2000) & (df_laptops['Price_euros'] <= 3000),
    (df_laptops['Price_euros'] > 800) & (df_laptops['Price_euros'] <=  2000),
    df_laptops['Price_euros'] <= 800
]

values = ['More expensive', 'Expensive', 'Cheap', 'Cheaper']
# >3000, >2000, >800, <= 800
```

```python
np.select(conditions, values)
```

```python
df_laptops['Price_levels'] = np.select(conditions, values)
df_laptops.head()
```

```python
df_laptops.value_counts('Price_levels')
```

```python
conditions = [
    df_laptops['Inches'] > 16,
    (df_laptops['Inches'] > 14) & (df_laptops['Inches'] <= 16),
    (df_laptops['Inches'] > 12) & (df_laptops['Inches'] <=  14),
    df_laptops['Inches'] <= 12
]

values = ['Bigger', 'Big', 'Small', 'Smaller']
# >16, >14, >12, <12
```

```python
df_laptops['screen_levels'] = np.select(conditions, values)
df_laptops.head()
```

```python
df_laptops.value_counts('screen_levels')
```
