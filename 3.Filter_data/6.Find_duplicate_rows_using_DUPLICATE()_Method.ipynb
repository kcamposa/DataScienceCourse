**EXERCISE #2**

```python
# order a dataframe asc by "Company" and 'Price'
# cheaper first and more expensive last
df_laptops = df_laptops.sort_values(['Company','Price_euros'], ascending=True)
df_laptops.head()
```

```python
# check all categories in the column 'Company'
df_laptops.value_counts('Company')
```

```python
# duplicate values in the column 'Company' default keep='first'
duplicated_first = df_laptops.duplicated('Company', keep='first')
```

```python
# show the dataframe with duplicated values in the column 'Company'

df_laptops[duplicated_first] # here we can see all duplicates


# keep='first' 
# show the dataframe with duplicated values in the column 'Company'
df_laptops[~duplicated_first] # the symbol ~ is the same of !, opposite of a value
df_laptops[~duplicated_first][['Company','Price_euros']]
# check all categories
#df_laptops[~duplicated_first].value_counts('Company')
```

```python
# keep='last'
duplicated_last = df_laptops.duplicated('Company', keep='last') # we get more expensive laptops
```

```python
# more expensive laptops by Company
df_laptops[~duplicated_last][['Company','Price_euros']] 
```

```python
# keep=False, meaning that every duplicate values doesnt appear
#show the dataframe without duplicates in the column Company
duplicated_false = df_laptops.duplicated('Company', keep=False) 
```

```python
df_laptops[~duplicated_false][['Company','Price_euros']] 
```
