```python
dir_datasets = '/content/drive/MyDrive/DataScience/Data Science Course/0.Datasets/{}'
df_laptops = pd.read_csv(dir_datasets.format('laptop_price.csv'), encoding='latin-1')
```

**CREATE A COLUMN BASED IN A CONDITION NP.WHERE()**

```python
# create an array based in prize column

np.where(df_laptops['Price_euros'] > 2000, 'Expensive', 'Cheap')
```

```python
# create a column
df_laptops['category_price'] = np.where(df_laptops['Price_euros'] > 2000, 'Expensive', 'Cheap')
```

```python
df_laptops.head()
```

```python
df_laptops.value_counts('category_price')
```
