```python
import pandas as pd
import numpy as np
```

```python
dir_datasets = '/content/drive/MyDrive/DataScience/Data Science Course/0.Datasets/{}'
df_laptops = pd.read_csv(dir_datasets.format('laptop_price.csv'), encoding='latin-1')
```

```python
df_laptops.head()
```

**DROP DUPLICATES()**

```python
# delete in 2 or more columns duplicate elements
df_laptops.drop_duplicates(['Company'])[['Company', 'Price_euros']]
```

```python
# order dataframe asc by company and price
# cheaper first and more expensive last
df_laptops = df_laptops.sort_values(['Company', 'Price_euros'], ascending=True)
```

```python
# cheaper laptops and dropping expensive laptops of the dataframe, keep='first'
df_laptops.drop_duplicates(['Company'], keep='first')[['Company', 'Price_euros']]
```

```python
# more expensive laptops and dropping cheaper laptops of the dataframe, keep='last'
df_laptops.drop_duplicates(['Company'], keep='last', inplace=True, ignore_index=True)
df_laptops[['Company', 'Price_euros']]
```
