# **CONCAT()**

VERTICAL CONCAT

```python
df1 = pd.DataFrame({'id' : ['A','B','C','D'],
                    'age' : [30, 23, 25, 22]})

df2 = pd.DataFrame({'id' : ['E','F','G','H'],
                    'age' : [40, 21, 19, 24]})
```

```python
df_concat = pd.concat([df1, df2], axis=0, ignore_index=True, sort=True)
df_concat
```
